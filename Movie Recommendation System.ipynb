```python
## Importing Modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Loading Movie Data Set
movies=pd.read_csv("movies.csv")
```

```python
movies.head()
```

```python
## Loading Ratings Data Set
ratings=pd.read_csv("ratings.csv")
```

```python
ratings.head()
```

```python
## Merging Data Set
df=pd.merge(movies, ratings)
```

```python
df.head()
```

```python
df = df.drop(['timestamp', 'genres', 'movieId'], axis=1)
```

```python
df.head()
```

```python
df1= df.pivot_table(index='userId', columns='title', values='rating')
```

```python
df1
```

```python
df1.shape
```

```python
## Feature Selection
df2=df1.dropna(thresh=15, axis=1).fillna(0)
```

```python
df2
```

```python
df2.shape
```

```python
## Correlation between Features
df3=df2.corr(method='pearson')
```

```python
df3
```

```python
# Movie recommendations
def recommendations(movie_name, ratings):
    similar_score = df3[movie_name] * (ratings - 2.5)
    similar_score = similar_score.sort_values(ascending=False)
    top_10_movies = similar_score.head(10)
    return top_10_movies.reset_index().rename(columns={movie_name: 'Score', 'index': 'title'})
```

```python
def display():
    # Ask the user for input
    movie_name = input("Enter the movie name: ")
    rating = float(input("Enter your rating for the movie (0-5): "))

    # Check if the movie is in the DataFrame
    if movie_name in df3.columns:
        # Get recommendations
        recommended_movies = recommendations(movie_name, rating)
        # Print the recommended movies
        print("Recommended movies based on your rating:\n")
        print(recommended_movies)
    else:
        print("Movie not found in the database. Please try again with a different movie.")
```

```python
display()
```
